```python
import numpy as np
import pandas as pd
```

```python
attraction_data = pd.read_csv("AttractionRawData.csv", index_col=0)
restaurant_data = pd.read_csv("RestaurantRawData.csv", index_col=0)
```

```python
# Get all unique tags
missing_tags_idx = []
food_tags = set()
attraction_tags = set()

def find_tags(tags, data: pd.DataFrame):
    for idx, row in data.iterrows():
        if type(row["tags"]) == str:
            if " • " in row["tags"]:
                sep = " • "
            else:
                sep = ","
            tag_str = row["tags"].split(sep)
            for tag in tag_str:
                tags.add(tag)
        else:
            missing_tags_idx.append(row["name"])

find_tags(food_tags, restaurant_data)
find_tags(attraction_tags, attraction_data)
```
