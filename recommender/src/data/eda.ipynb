```python
import numpy as np
import pandas as pd
```

```python
attraction_data = pd.read_csv("AttractionRawData.csv", index_col=0)
restaurant_data = pd.read_csv("RestaurantRawData.csv", index_col=0)
```

```python
# Before removing duplicates
print(f"{len(attraction_data)=}")
print(f"{len(restaurant_data)=}")
```

```python
# Drop duplicates
attraction_data = attraction_data.drop_duplicates()
restaurant_data = restaurant_data.drop_duplicates()
```

```python
# After removing duplicates
print(f"{len(attraction_data)=}")
print(f"{len(restaurant_data)=}")
```

```python
# Get all unique tags and identify problematic data points
missing_tags_food = []
missing_tags_attraction = []
food_tags = set()
attraction_tags = set()
```

```python
def find_tags(tags: set, missing_tags: list, data: pd.DataFrame):
    for idx, row in data.iterrows():
        if type(row["tags"]) == str:
            if " • " in row["tags"]:
                sep = " • "
            else:
                sep = ","
            tag_str = row["tags"].split(sep)
            for tag in tag_str:
                tags.add(tag)
        else:
            missing_tags.append(row["name"])
```

```python
find_tags(food_tags, missing_tags_food, restaurant_data)
```

```python
find_tags(attraction_tags, missing_tags_attraction, attraction_data)
```

```python
with open("food_tags.txt", "w") as f:
    for f_tag in food_tags:
        f.write(f_tag + "\n")
```

```python
with open("attraction_tags.txt", "w") as f:
    for a_tag in attraction_tags:
        f.write(a_tag + "\n")
```
